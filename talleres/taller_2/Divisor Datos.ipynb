## Divisor
Cuaderno que divide las calificaciones por grupo para facilidad de procesamiento

```python
import json
import glob
import pandas as pd
import os


files = glob.glob('data_raw/CF_base_formato_json/*.json')
final_files = list(files)

num_modules = 20

def dar_grupo_usuario(s , num_modules = 20):
    return(abs(ord(s[0]) + ord(s[1]) + ord(s[-1])) % (num_modules))
```

```python
# Group IDS Construction
for m in range(num_modules):
    
    print('Started With Group: {} of {}'.format(m +1, num_modules))
        
    current_ids = set()
    i = 0
    for file in final_files:
        
        with open(file) as f:
            lines = []
            for jsonObj in f:
                lines.append(json.loads(jsonObj))
                
            dataframe_row = pd.DataFrame(lines)[['user_id']]
            current_ids = current_ids.union(dataframe_row.loc[dataframe_row.user_id.apply(lambda s : dar_grupo_usuario(s)) == m].user_id.values)            
    
        if (i % 10) == 0:
            print('{} of {} Current Users: {}'.format(i + 1, len(final_files), len(current_ids)))
            
        i += 1
    
    pd.DataFrame({'user_id': list(current_ids)}).to_csv('data/id_groups/ids_{}.csv'.format(m), index = False)
    print()
    print('Finished With Group: {} of {}'.format(m +1, num_modules))
    print()
```

```python
# Crea los encabezados
encabezado = 'user_id,business_id,year_review,review_id,stars,last_review'
for m in range(num_modules):
    with open('data_raw/CF_base_formato_json_grouped/reviews_{}.csv'.format(m), 'w') as f:
        f.write(encabezado + '\n')
```

```python
for k in range(len(final_files)):
    
    print('Starts with File: {} of {}'.format(k + 1, len(final_files)))
    file = final_files[k]
    # Loads File
    with open(file) as f:

        lines = []
        for jsonObj in f:
            lines.append(json.loads(jsonObj))
            
    df = pd.DataFrame(lines)
    
    for m in range(num_modules):
        
        print('   Group: {} of {}'.format(m + 1, num_modules))
        
        # Loads Ids
        current_ids = pd.read_csv('data/id_groups/ids_{}.csv'.format(m))
        
        # Filters
        df_filtered = df.merge(current_ids, on = 'user_id')
        
        # Saves the file
        df_filtered.to_csv('data_raw/CF_base_formato_json_grouped/reviews_{}.csv'.format(m), mode='a', header=False, index = False)
    
    print()
    print()
print('------------------')
print('Done')
print('------------------')
```

```python
# Converts to pickle
for m in range(num_modules):
    print('Compressing: {} of {}'.format(m +1, num_modules))
    
    # Imports
    org_file = 'data_raw/CF_base_formato_json_grouped/reviews_{}.csv'.format(m)
    df = pd.read_csv(org_file)
    
    # Compress
    df.to_pickle('data_raw/CF_base_formato_json_grouped/reviews_{}.zip'.format(m))
    
    # Removes
    os.remove(org_file)
    
print('Done')
```
